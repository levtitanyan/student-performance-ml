```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features)", fontsize=14, weight="bold")
plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm

# Drop non-numeric columns for OLS (handle categoricals later if needed)
X = data.select_dtypes(include=["int64", "float64"]).drop(columns=["G3"])
y = data["G3"]

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
